```python
# dtype()
import numpy as np
arr1 = np.array([1, 2, 3])
print("arr1.item:", arr1.itemsize)
print("arr1.dtype >", arr1.dtype, arr1.nbytes, "\n")

arr2 = np.array([1, 2, 3], dtype='int32')
print("arr2.item:", arr2.itemsize)
print("arr2.dtype >", arr2.dtype, arr2.nbytes, "\n")

arr3 = np.ones((1000))
print(arr3.nbytes)
arr4 = np.ones((1000), dtype='int32')
print(arr4.nbytes)
```

```python

# sum() yigindi ,  mean() ortachasini aniqlaydi,   max() arraydagi maksimal ,  min() arraydagi minimal
arr = np.array([12, 2, 3, 7, 5, 7])

print("sum:", arr.sum())
print("mean:", arr.mean())
print("max:", arr.max())
print("min:", arr.min())
```

```python
# flatten() 2d yoki 3d arrayni 1d ga ozgartiradi

arr = np.array([[12, 2, 3], [1, 2, 5]])
print("flatten:", arr.flatten())
```

```python

# T - transpose
arr1 = np.array([[12, 2, 3, 4], [1, 2, 5, 8]])
print("arr1.transpose:\n", arr1.T, "\n")

arr2 = np.array([
    [[12, 2, 3, 4], [1, 2, 5, 8]],
    [[17, 5, 2, 4], [8, 7, 5, 3]],
    [[7, 15, 0, 2], [3, 1, 9, 1]],
  ])
print("arr2.transpose:\n", arr2.transpose(1,0,2))
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[2, 0], [5, 7]])

# linier kopaytirish
multip1 = arr1 @ arr2
print("multip1:", multip1, "\n")

# dot function
multip2 = np.dot(arr1, arr2)
print("multip2:", multip2, "\n")

# dot method
multip3 = arr1.dot(arr2)
print("multip3:", multip3, "\n")

# haqiqiy kopaytirish emas!
multip4 = arr1 * arr2
print("multip4:", multip4)
```
